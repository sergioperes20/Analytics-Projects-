```python
import pandas as pd

file_path = "Eletric_Cars_US.xlsx"
file_world = "Eletric_Cars_World.xlsx"

eletric_file_US = pd.read_excel(file_path)
eletric_file_World = pd.read_excel(file_world)
```

```python
print(eletric_file_US.describe())
```

```python
print(eletric_file_World.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st

correlation_matrix = eletric_file_US.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
eletric_world_2021 = eletric_file_World[eletric_file_World['year'] == 2021]


plt.figure(figsize=(8, 8))
sns.barplot(x='value', y='region', data=eletric_world_2021, palette='Set3')
plt.xlabel('Value')
plt.ylabel('Region')
plt.title(' Region/Value in 2021')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(data=eletric_file_US, x='State')
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Count of Electric Cars by State')
plt.xticks(rotation=45)
plt.show()
```

```python
mean1 = eletric_file_US['Electric Vehicle (EV) Total']
mean2 = eletric_file_US['Battery Electric Vehicles (BEVs)']

t_statistic, p_value = st.ttest_ind(mean1, mean2, equal_var=True)

alpha =0.05

if p_value < alpha:
    print(f"Reject the null hypothesis. There is a statistically significant difference between {mean1.name} and {mean2.name}.")
else:
    print(f"Fail to reject the null hypothesis. There is no statistically significant difference between {mean1.name} and {mean2.name}.")

```
